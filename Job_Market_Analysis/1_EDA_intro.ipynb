# Exploratory Data Analysis

### Roles to Explore: 

```python
# importing libraries 
import ast
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
#loading data
df = pd.read_csv('data_jobs.csv') 
# data cleanup 
df['job_posted_date']= pd.to_datetime(df['job_posted_date']) # converting the job_posted_date column to a datetime object
df['job_skills'] = df['job_skills'].apply(lambda skill_list: ast.literal_eval(skill_list) if pd.notna(skill_list) else skill_list) #changing the string representation of a list to an actual list in this column 
```

### Filter for US Data Analyst roles

```python
#viewing our dataframe before filtering 
df
```

```python
# filtering where job_country = United States and job_title_short = Data Analyst 
# then converting it into a new dataframe
df_da_us = df[(df['job_country'] == 'United States') & (df['job_title_short'] == 'Data Analyst')]
# viewing dataframe to pick some columns for EDA 
df_da_us
```

### Lets look at the top locations for Data Analysts in the United States

```python
# We are grabbing the job location column from our data and running a count of values, then getting the first 10. Then assigning to a dataframe.
df_da_us['job_location'].value_counts().head(10).to_frame()
# lets assign this to a variable for plotting 
df_da_us_plot = df_da_us['job_location'].value_counts().head(10).to_frame()
#view the new data set 
df_da_us_plot
```

```python
# Creating a barplot using seaborn where paramaters (data, x, y) are taken in 
sns.barplot(df_da_us_plot, x='count', y=df_da_us_plot.index, palette='dark:b')
sns.despine() # removes box border around the original plot 
plt.ylabel('Location') # y-axis title
plt.xlabel('Job Postings') # x-axis title
plt.title('Count of Job Locations for Data Analyst in the US') #main title
plt.show()

# note: Anywhere must be correlated with WFH(Working from Home)
```

# Lets check some specifics for Data Analyst roles in the United States

### Our data frame has info such as "Work From Home", 'No Degree Mentioned', and 'Health Insurance'  within each posting sorted as boolean values. 
### Lets Visualize these columns. 

```python
# viewing the dataframe to correctly pick out these columns
df_da_us
```

```python
# grabbing the columns we need, and storing them in a dictionary for iterative graphing 
dict_values = {
    'job_work_from_home': 'Work From Home', 
    'job_no_degree_mention': 'Degree Requirement', 
    'job_health_insurance': 'Health Insurance'
}
# the reason for creating this dictionary is to be able to iterate the key 'columns' and output the value pair as the 'Title' in our graph. 
```

```python
# lets now create a figure of piecharts for each key column
fig, ax = plt.subplots(1, 3) # one row with 3 graphs

# lets enumerate our dictionary for plotting 
for i, (column, title) in  enumerate(dict_values.items()):
    ax[i].pie(df_da_us[column].value_counts(), labels=['False', 'True'], autopct='%1.1f%%', startangle=90) # for each value at index [i], plot a pie chart, and format to a percentage. 
    ax[i].set_title(title) # for each index row, return the value from the dictionary as a title
plt.suptitle
plt.show()
```

# Lastly, Lets explore the top companies that are in need for Data Analysts in the US

```python
# grabbing the 10 ten companies by count of job postings and setting it to a data frame 
df_us_companies = df_da_us['company_name'].value_counts().head(10).to_frame()
df_us_companies
```

```python
# Creating a barplot using seaborn with our df_us_companies dataframe
sns.barplot(data= df_us_companies, x='count', y=df_us_companies.index, palette='dark:b')
sns.despine() # removes box border around the original plot 
plt.ylabel('') # y-axis title
plt.xlabel('Job Postings') # x-axis title
plt.title('Count of Job Postingsfor Data Analyst in the US by Company') #main title
plt.show()
```
